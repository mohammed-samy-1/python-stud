```python
import nltk
```

The code builds a Conditional Frequency Distribution
That counts the occurrences of specified target words ('men', 'women', 'people') in the State of the Union addresses, categorized by the first four characters of the file identifier (year).

```python
cfd = nltk.ConditionalFreqDist(
  (target, fileid[:4])
  # This iterates over each file identifier in the state of the union addresses.
  for fileid in nltk.corpus.state_union.fileids()
  # This nested loop iterates over each word in the current state of the union address.
  for w in nltk.corpus.state_union.words(fileid)
  # This is another nested loop that iterates over the specified target words ('men', 'women', 'people')
  for target in ['men', 'women', 'people']
  if w.lower().startswith(target))
```

### draw the change of the words (men, women, people) appearance piquancy over years

```python
cfd.plot()
```
